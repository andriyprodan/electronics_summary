```python
import numpy as np
import matplotlib.pyplot as plt

# Time axis
f = 60.0                 # Hz
w = 2 * np.pi * f        # rad/s
t = np.linspace(0, 2/f, 2000)  # two periods

A = 1.0  # amplitude

# Three sinusoids (typical phase relationships for a sinusoidal excitation):
# - Pure resistance: in phase (0°)
# - Pure capacitive reactance: current leads voltage by +90°
# - Pure inductive reactance: current lags voltage by -90°
y_R = A * np.sin(w * t + 0.0)
y_XC = A * np.sin(w * t + np.pi/2)
y_XL = A * np.sin(w * t - np.pi/2)

plt.figure(figsize=(10, 4))
plt.plot(t, y_R, label='Resistance (0°)', linewidth=2)
plt.plot(t, y_XC, label='Capacitive reactance (+90°)', linewidth=2)
plt.plot(t, y_XL, label='Inductive reactance (-90°)', linewidth=2)

plt.title('Three sinusoids: R, Xc, Xl')
plt.xlabel('Time, s')
plt.ylabel('Amplitude')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
